##### –°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –∂–∞–Ω—Ä—ã (—Ç–æ–ø-5). –ü–æ—è—Å–Ω–∏—Ç–µ —Ä–∞–∑–Ω–∏—Ü—É.

```python
for name in sales_region: #—Ü–∏–∫–ª–æ–º –ø–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º –ø–æ –∂–∞–Ω—Ä–∞–º –ø–æ –∫–∞–∂–¥–æ–º—É —Ä–µ–≥–∏–æ–Ω—É
    top_sales_genre.pivot_table(index='genre', values=name, aggfunc='sum').sort_values(by=name, ascending=False
                                        ).head().plot(kind='pie', y=name, legend=False, figsize=(7,7), fontsize=14, autopct='%1.1f%%')
    plt.title(f'{name}', fontsize=16)
    plt.ylabel('');
```
