````python
def floor_number(row):  #–Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã, –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ floor_number
    if row['floor'] == 1:
        return '–ø–µ—Ä–≤—ã–π'
    if row['floor'] == row['floors_total']:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    return '–¥—Ä—É–≥–æ–π'

data['floor_number'] = data.apply(floor_number, axis=1) #–≤—ã–∑–æ–≤–µ–º —Ñ—É–Ω–∫—Ü–∏—é
````
